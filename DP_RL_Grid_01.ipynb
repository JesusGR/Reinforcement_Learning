```python
import numpy as np
# Entrada: numpy array con la funciÃ³n de valor para cada estado
def policyEvaluation(grid):
    #necesitamos una copia del grid para actualizar los valores
    new_values=np.copy(grid)
    for i in range(grid.shape[0]):
        
        for j in range(grid.shape[0]):
            vecinos=[]
            # Si son terminales no se realiza nada
            if (i==0 and j==0) or (i==grid.shape[0]-1 and j==grid.shape[0]-1):
                pass
            # PolÃ­tica aleatoria (todos con las mismas probabilidades 0.25)
            # caso 1: esquina inferior izquierda
            elif i+1==grid.shape[0] and j-1<0:
                vecinos.append(grid[i][j])
                vecinos.append(grid[i][j])
                vecinos.append(grid[i][j+1])
                vecinos.append(grid[i-1][j])
            #caso 2: esquina superior derecha
            elif j+1==grid.shape[0] and i-1<0:
                vecinos.append(grid[i][j])
                vecinos.append(grid[i][j])
                vecinos.append(grid[i][j-1])
                vecinos.append(grid[i+1][j])
            #caso 3: fila superior
            elif i-1<0:
                vecinos.append(grid[i][j])
                vecinos.append(grid[i][j-1])
                vecinos.append(grid[i][j+1])
                vecinos.append(grid[i+1][j])
            #caso 4: fila inferior
            elif i+1==grid.shape[0]:
                vecinos.append(grid[i][j])
                vecinos.append(grid[i-1][j])
                vecinos.append(grid[i][j-1])
                vecinos.append(grid[i][j+1])
            #caso 5: columna izquierda
            elif j-1<0:
                vecinos.append(grid[i][j])
                vecinos.append(grid[i-1][j])
                vecinos.append(grid[i+1][j])
                vecinos.append(grid[i][j+1])
            #caso 6: columna derecha
            elif j+1==grid.shape[0]:
                vecinos.append(grid[i][j])
                vecinos.append(grid[i-1][j])
                vecinos.append(grid[i+1][j])
                vecinos.append(grid[i][j-1])
            #caso 7: todos los demas
            else:
                vecinos.append(grid[i+1][j])
                vecinos.append(grid[i-1][j])
                vecinos.append(grid[i][j-1])
                vecinos.append(grid[i][j+1])
            n_v=0.0
            
            if len(vecinos)>0:
                for k in vecinos:
                    n_v+=(0.25*(-1.0+k))
                new_values.itemset((i,j),n_v)
    #print(grid)
    return new_values
v=[]
v.append(np.zeros((4,4),dtype=float))
print(v[-1])
for i in range(11):
    print('k='+str(i))
    v.append(policyEvaluation(v[-1]))
    print(np.around(v[-1], decimals=4))
#grid_1=
#print(grid_1)
#grid_2=policyEvaluation(grid_1)
#print(grid_2)
#grid_3=policyEvaluation(grid_2)
#print(grid_3)

```
